<h1><center> Misc Info </center></h1>

File name format: \<ADL OR FALL_CODE>_\<SUBJECT_ID>_\<TRIAL_NO>.txt

Sensors:

Data were acquired with three sensors (2 accelerometers and 1 gyroscope) at a frequency sample of 200 Hz.

Each file contains nine columns and a different number of rows depending on the test length.

1st column is the acceleration data in the X axis measured by the sensor ADXL345.
2nd column is the acceleration data in the Y axis measured by the sensor ADXL345.
3rd column is the acceleration data in the Z axis measured by the sensor ADXL345.

4th column is the rotation data in the X axis measured by the sensor ITG3200.
5th column is the rotation data in the Y axis measured by the sensor ITG3200.
6th column is the rotation data in the Z axis measured by the sensor ITG3200.

7th column is the acceleration data in the X axis measured by the sensor MMA8451Q.
8th column is the acceleration data in the Y axis measured by the sensor MMA8451Q.
9th column is the acceleration data in the Z axis measured by the sensor MMA8451Q.

Data are in bits with the following characteristics:

ADXL345:
Resolution: 13 bits
Range: +-16g

ITG3200
Resolution: 16 bits
Range: +-2000°/s

MMA8451Q:
Resolution: 14 bits
Range: +-8g


In order to convert the acceleration data (AD) given in bits into gravity, use this equation: 

Acceleration $g = [(2*Range)/(2^{Resolution})]*AD$

In order to convert the rotation data (RD) given in bits into angular velocity, use this equation:

Angular velocity $°/s = [(2*Range)/(2^{Resolution})]*RD$

# Data Processing + Feature Engineering

```python
# Process raw data. SisFall has 2 accelerometers, find their average
# Also convert accel into in terms of g, and rotation into angular velocity degrees/seconds
def xyz_accel_rota(original_df):
    df = original_df.copy()
    df[8] = (df[8].str.replace(';', '')).astype(float)
    accel_cols = [0,1,2,6,7,8]
    for i in accel_cols:
        if i < 3: # ADXL345 has resolution 13bits, range +/-16g
            df[i] = ((2*16)/(2**13))*df[i] # Convert to acceleration, in g
        elif i > 5: # MMA8451Q has resolution 14bits, range +/-8g
            df[i] = ((2*8)/(2**14))*df[i]
    rota_cols = [3,4,5]
    for j in rota_cols: # ITG3200 has resolution 16bits, range +- 2000 deg/s
        df[j] = ((2*2000)/(2**16))*df[j]
    
    df['x_accel_g'] = (df[0]+df[6])/2
    df['y_accel_g'] = (df[1]+df[7])/2
    df['z_accel_g'] = (df[2]+df[8])/2
    
    df['net_acceleration'] = np.sqrt(df['x_accel_g']**2 + df['y_accel_g']**2 + df['z_accel_g']**2)
    
    df.rename({3:'x_rota',4:'y_rota',5:'z_rota'},axis=1,inplace=True)
    df.drop([0,1,2,6,7,8],axis=1,inplace=True) # Just use the averaged accel
    return(df)

# Feature engineering
# Each SisFall data .txt file has a few thousand rows. This compresses each file into a single row, for ML
# (min/max/mean/var/skew/kurtosis) * ( (x/y/z) * (accel/rota) + net_accel) = 42 columns
def create_features(df):
    Min = df.min(axis=0).reset_index().rename({0:'min'},axis=1)
    Max = df.max(axis=0).reset_index().rename({0:'max'},axis=1)
    Mean = df.mean(axis=0).reset_index().rename({0:'mean'},axis=1)
    Var = df.var(axis=0).reset_index().rename({0:'variance'},axis=1)
    Skew = df.skew(axis=0).reset_index().rename({0:'skewness'},axis=1)
    Kurtosis = df.kurtosis(axis=0).reset_index().rename({0:'kurtosis'},axis=1)
    
    df_features = reduce(lambda left,right: pd.merge(left,right,on='index',)
                         ,[Min,Max,Mean,Var,Skew,Kurtosis])
    
    df_features.set_index('index',inplace=True)
    df_features = df_features.unstack().to_frame().sort_index(level=1).transpose()
    df_features.columns = df_features.columns.map('_'.join)

    return(df_features)

# For reading all files in the SisFall dataset
def read_files(readDir = Path("D:\\Project Morphine\\SisFall_dataset")):
    df_list = []
    for folder_name in os.listdir(readDir):
#         print(folder_name)
        if folder_name == 'Readme.txt': # skip this file
            continue
        for file in os.listdir(readDir/folder_name):
            if file.endswith('.txt'): # have some random files which need to be skipped
#                 df_list.append(pd.read_csv(readDir/folder_name/file,header=None)) # Read file itself
#         df = pd.concat(df_list,axis=0,ignore_index=True)
                df_list.append(readDir/folder_name/file) # Just file paths/names
    return(df_list)
```

```python
all_file_names = read_files()

# Initialise overall dataframe with first datafile
initial_df = pd.read_csv(all_file_names[0],header=None)
initial_df = xyz_accel_rota(initial_df)
overall_df = create_features(initial_df)

# Get filename, add it as a new column
details = str(all_file_names[0]).rsplit('\\',1)[1]
details = details.replace('.txt','')

overall_df['details'] = details
del initial_df

# Iterate for all remaining .txt data file in SisFall dataset
for i in all_file_names[1:]:
    # Initialise overall dataframe
    df = pd.read_csv(i,header=None)
    df = xyz_accel_rota(df)
    df = create_features(df)

    # Get filename, add it as a new column
    details = str(i).rsplit('\\',1)[1]
    details = details.replace('.txt','')
    df['details'] = details

    overall_df = pd.concat([overall_df,df],axis=0)
    
# Find details based on filename
overall_df[['activity_type','subject_id','trial_number']] = overall_df['details'].str.split('_',expand=True)

# True if activity is a fall, False if it's an ADL
overall_df['fall_bool'] = overall_df['activity_type'].str.contains('F')

# elderly: between 60,75 years old. Not elderly: between 19,30 years old.
overall_df['elderly_bool'] = overall_df['subject_id'].str.contains('SE')

overall_df = overall_df.merge(SUBJECTS,how='left',on='subject_id')

overall_df['male_bool'] = np.where(overall_df['Gender'] == 'M',True,False)

overall_df.drop(['details','trial_number','Gender'],axis=1,inplace=True)

overall_df.head()
```
